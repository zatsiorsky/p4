```python
import os.path
import pandas as pd
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('grid_results_final.csv')
```

```python
df.columns = df.columns.str.lstrip()
```

```python
font = {'color':  'red',
        'size': 14,
        }

plt.figure(figsize=[10,5])
plt.title("Mean Score Over 1000 Iterations of 100 Games", size=15)
plt.scatter(df.index, df['mean'], alpha = .3, label="100 Games Iteration Mean")
plt.hlines(df['mean'].mean(), -10, 1010, label= 'Mean Across All Iterations', 
           linestyles='dashed', colors='r', linewidth=2)
plt.xlabel('Iteration of 100 Games')
plt.ylabel('Mean Score')
plt.text(1020, df['mean'].mean(), str(df['mean'].mean()), fontdict=font)
plt.xlim([-10,1100])
plt.legend()
plt.savefig('../figures/mean_scores_over_iters.png');
```

```python
font = {'color':  'red',
        'size': 14,
        }

plt.figure(figsize=[10,5])
plt.title("Max Score Over 1000 Iterations of 100 Games", size=15)
plt.scatter(df.index, df['max'], alpha = .3, label="100 Games Iteration Max")
plt.hlines(df['max'].mean(), -10, 1010, label= 'Mean of Max Across All Iterations', 
           linestyles='dashed', colors='r', linewidth=2)
plt.xlabel('Iteration of 100 Games')
plt.ylabel('Max Score')
plt.text(1020, df['max'].mean(), str(df['max'].mean()), fontdict=font)
plt.xlim([-10,1100])
plt.legend()
plt.savefig('../figures/max_scores_over_iters.png');
```

```python
font = {'color':  'red',
        'size': 14,
        }

plt.figure(figsize=[8,4])
plt.title("Distribution of Mean Score Over 1000 Iterations of 100 Games", size=12)
sns.distplot(df['mean'].values, label="100 Games Iteration Mean", norm_hist=False)
#plt.hlines(, -10, 1010, label= 'Mean Across All Iterations', 
#           linestyles='dashed', colors='r', linewidth=2)
plt.xlabel('Mean Score', size=12)
plt.ylabel('Frequency', size=12)
plt.vlines(df['mean'].mean(), 0,0.0095, linestyles='dashed', colors='r', linewidth=2, alpha=.5)
plt.text(df['mean'].mean(), 0.0095, "{:,}".format(df['mean'].mean()), fontdict=font)
plt.savefig('../figures/mean_dist.png');
```

```python
font = {'color':  'red',
        'size': 14,
        }

plt.figure(figsize=[8,4])
plt.title("Distribution of Max Score Over 1000 Iterations of 100 Games", size=12)
sns.distplot(df['max'].values, norm_hist=False)
#plt.hlines(df['mean'].mean(), -10, 1010, label= 'Mean Across All Iterations', 
#           linestyles='dashed', colors='r', linewidth=2)
plt.xlabel('Max Score', size=12)
plt.ylabel('Frequency', size=12)
plt.vlines(df['max'].mean(), 0,0.00085, linestyles='dashed', colors='r', linewidth=2, alpha=.5)
plt.text(df['max'].mean(), 0.00085, "{:,}".format(df['max'].mean()), fontdict=font)
#plt.text(1020, df['mean'].mean(), str(df['mean'].mean()), fontdict=font)
plt.savefig('../figures/max_dist.png');
```
